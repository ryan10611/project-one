```python
import pandas as pd
import matplotlib as plt
import requests
import json
import os
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
data = 'finaltable.csv'
df = pd.read_csv(data)
```

```python
df.head()
```

```python
df['intercept'] = 1
X = df[['intercept', 'POPULATION', 'LAND AREA (mi^2)', '# PHARMACIES']]
y = df['AVERAGE LIFE EXPECTANCY']

smresults = sm.OLS(y, X).fit()
df['pred'] = smresults.predict()

print(smresults.summary())
```
